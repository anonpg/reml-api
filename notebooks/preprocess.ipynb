```python
df = pd.read_pickle('/tmp/raw.pkl')

df = df.drop(columns=['No'])

df = df.rename(columns={
    '種類': 'type',
    '地域': 'zone_type',
    '土地の形状': 'shape',
    '今後の利用目的': 'future_purpose',
    '前面道路：方位': 'front_road_dir',
    '前面道路：種類': 'front_road_type',
    '前面道路：幅員（ｍ）': 'front_road_width',
    '都市計画': 'city_plan',
    '建ぺい率（％）': 'building_cov',
    '容積率（％）': 'floor_cov',
    '改装': 'renovation',
    
    '市区町村コード': 'city_code',
    '都道府県名': 'prefecture',
    '市区町村名': 'city',
    '地区名': 'city2',
    '最寄駅：名称': 'nearest_sta',
    '最寄駅：距離（分）': 'nearest_sta_dist',
    '取引価格（総額）': 'price',
    '坪単価': 'tubo_price',
    '取引価格（㎡単価）': 'm2_price',
    '間取り': 'layout',
    '面積（㎡）': 'area',
    '間口': 'frontage',
    '延床面積（㎡）': 'floor_area',
    '建築年': 'building_year',
    '建物の構造': 'structure', # multiple value
    '用途': 'purpose', # multiple value
    '取引時点': 'time',
    '取引の事情等': 'notes', # multiple value
})

df['nearest_sta_dist'] = df['nearest_sta_dist'].replace('30分?60分', '45').replace('1H?1H30', '75').replace('1H30?2H', '105').replace('2H?', '120').astype('float')
df['area'] = df['area'].replace(['2000㎡以上', '2,000m&sup2;以上'], '2000').replace(['5000㎡以上', '5,000m&sup2;以上'], '5000').astype('int')
df['frontage'] = df['frontage'].replace('50.0m以上', '50').astype('float')
df['floor_area'] = df['floor_area'].replace('10m^2未満', '5').replace('2000㎡以上', '2000').astype('float')

def gengou_to_seireki(x):
    if isinstance(x, float):
        return x
    
    x = x.replace('年', '')
        
    if x == '戦前':
        return 1945
    elif '昭和' in x:
        return int(x.replace('昭和', '')) + 1925
    elif '平成' in x:
        return int(x.replace('平成', '')) + 1988
    elif '令和' in x:
        return int(x.replace('令和', '')) + 2018
    
    return x

df['building_year'] = df['building_year'].apply(gengou_to_seireki).astype('float')

def transform_time(x):
    return (x // 10) * 100 + (x % 10) * 25

df['time'] = df['time'].str.replace('年第', '').str.replace('四半期', '').astype('int').apply(transform_time)

print(df['time'].unique())
    
print(df.dtypes)

df = df.sort_values(['time'], kind='stable')
df = df.reset_index(drop=True)
df.to_pickle('/tmp/all.pkl')
display(df)
```

```python
df = pd.read_pickle('/tmp/all.pkl')

df = df.loc[df['prefecture'] == '東京都']

df_pos = pd.read_pickle('/tmp/pos_all.pkl')
df_pos = df_pos.rename(columns={
    '市区町村コード': 'city_code',
    '大字町丁目名': 'city2',
    '緯度': 'latitude', 
    '経度': 'longitude',
})
df = df.merge(df_pos[['city_code', 'city2', 'latitude', 'longitude']].set_index(['city_code', 'city2']), how='left', on=['city_code', 'city2'])
# df = df.merge(df_pos[['city_code', 'latitude', 'longitude']].groupby(['city_code']).mean(), how='left', on=['city_code'])

df_pos2 = df_pos.copy()
# df_pos2['city2'] = np.nan
df_pos2['city2'] = df_pos2['city2'].str.replace('一丁目', '')
df2 = df.merge(df_pos2.groupby(['city_code', 'city2'])[['latitude', 'longitude']].mean(), how='left', on=['city_code', 'city2'])
df.loc[df['latitude'].isnull(), 'latitude'] = df2.loc[df['latitude'].isnull(), 'latitude_y']
df.loc[df['longitude'].isnull(), 'longitude'] = df2.loc[df['longitude'].isnull(), 'longitude_y']

print(df.isnull().sum())

df.to_pickle('/tmp/all_with_pos.pkl')
display(df)
```
